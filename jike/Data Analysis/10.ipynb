```python
"""刘亦菲一共10723张图片，写个for循环跑完所有请求"""
#coding:utf-8
import requests
import json
query='刘亦菲'
"""下载图片"""
def download(src,id):
    dir='D:/代码/jike/Data Analysis/刘亦菲'+str(id)+'.jpg'
    #dir = './'+str(id)+'.jpg'
    try:
            pic = requests.get(src,timeout=10)
            fp = open(dir,'wb')
            fp.write(pic.content)
            fp.close()
    except requests.exceptions.ConnectionError:
            print('图片无法下载')
            
```

```python
def searchImages():
    """for 循环 请求全部的url"""
    for i in range(0,10723,20):#一次只能请求20张
        url='https://www.douban.com/search?cat=1025&q='+query+'&limit=20&start='+str(i)
        html = requests.get(url).text  #得到返回结果
        print('html:'+html)
        response = json.loads(html,encoding='utf-8')  #将JSON格式转换为Python对象

        for image in response['images']:
            print(image['src'])  #查看当前下载的图片网址
            download(image['src'],image['id'])  #下载一张图片
```

```python
searchImages()
```
